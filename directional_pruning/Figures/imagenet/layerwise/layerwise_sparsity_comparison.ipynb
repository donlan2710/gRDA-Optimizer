```python
df_all = pd.read_csv("all_comparison.csv")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
df_all_y_inverse = df_all.copy()
```

```python
for col in df_all_y_inverse.columns[3:]: 
    df_all_y_inverse[col] = 1-df_all_y_inverse[col]
```
